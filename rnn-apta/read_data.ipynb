```python

# Helper function to process and print results
def process_and_print(data, day):
    # Clean column names
    data.columns = data.columns.str.strip().str.replace(' ', '_')

    # Encode labels: Benign = 0, others = 1
    data['Label'] = data['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)

    data['Protocol'] = pd.to_numeric(data['Protocol'], errors='coerce')
    data = data[data['Protocol'] == 6]

    # Filter where Source Port or Destination Port is 80
    filtered = data[(data['Source_Port'] == 80) | (data['Destination_Port'] == 80)]
    filtered2 = data[(data['Source_Port'] == 443) | (data['Destination_Port'] == 443)]
    filtered3 = data[(data['Source_Port'] == 21) | (data['Destination_Port'] == 21)]
    filtered4 = data[(data['Source_Port'] == 22) | (data['Destination_Port'] == 22)]
    filtered5 = data[(data['Source_Port'] == 465) | (data['Destination_Port'] == 465)]

    # Count the labels
    counts = filtered['Label'].value_counts()
    counts2 = filtered2['Label'].value_counts()
    counts3 = filtered3['Label'].value_counts()
    counts4 = filtered4['Label'].value_counts()
    counts5 = filtered5['Label'].value_counts()

    # Print result
    print(f"--- {day} ---")
    print(f"Port 80 =\n{counts}", "\n")
    print(f"Port 443 =\n{counts2}", "\n")
    print(f"Port 21 =\n{counts3}", "\n")
    print(f"Port 22 =\n{counts4}", "\n")
    print(f"Port 465 =\n{counts5}", "\n")

# Process each day
process_and_print(data1, "Monday")
process_and_print(data2, "Tuesday")
process_and_print(data3, "Wednesday")
process_and_print(data4, "Thursday")
process_and_print(data5, "Friday")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# Load the CSV files
df1 = pd.read_csv('/Users/aptar/OneDrive/Documents/AA_KULIAH/TA/neuralnetwork-AD/rnn-apt/errors/testing_errors_Tuesday-WorkingHours_testing_22.csv')
df2 = pd.read_csv('/Users/aptar/OneDrive/Documents/AA_KULIAH/TA/neuralnetwork-AD/rnn-apt/errors/training_errors_Monday-WorkingHours_training_22.csv')

# Extract the floating prediction error column
column_2_values1 = df1['binary_prediction_error']
column_2_values2 = df2['binary_prediction_error']
# column_2_values1 = df1['floating_prediction_error']
# column_2_values2 = df2['floating_prediction_error']

# Plot the two distributions
plt.figure(figsize=(10, 6))
plt.hist(column_2_values1, bins=50, alpha=0.6, label='Test Errors', color='skyblue', density=True,)
plt.hist(column_2_values2, bins=50, alpha=0.6, label='Train Errors', color='orange', density=True)
plt.title('Distribution of Floating Prediction Errors')
plt.xlabel('Prediction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Drift detection function
def detect_drift(train_errors, test_errors, alpha=0.05):
    stat, p_value = ks_2samp(train_errors, test_errors)
    drift_detected = p_value < alpha
    return drift_detected, {"statistic": stat, "p_value": p_value}

# Call the function
drift, result = detect_drift(column_2_values2, column_2_values1)

print("Drift Detected:", drift)
print("KS Test Statistic:", result["statistic"])
print("P-value:", result["p_value"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV files
df1 = pd.read_csv('/Users/aptar/OneDrive/Documents/AA_KULIAH/TA/neuralnetwork-AD/rnn-apt/errors/old/testing_errors_Tuesday-WorkingHours_testing_80.csv')
df2 = pd.read_csv('/Users/aptar/OneDrive/Documents/AA_KULIAH/TA/neuralnetwork-AD/rnn-apt/errors/training_errors_Monday-WorkingHours_training_80.csv')


# Extract the floating prediction error column
column_2_values1 = df1['binary_prediction_error']
column_2_values2 = df2['binary_prediction_error']
# column_2_values1 = df1['floating_prediction_error']
# column_2_values2 = df2['floating_prediction_error']

# Plot the two distributions
plt.figure(figsize=(10, 6))
plt.hist(column_2_values1, bins=50, alpha=0.6, label='Test Errors', color='skyblue', density=True)
plt.hist(column_2_values2, bins=50, alpha=0.6, label='Train Errors', color='orange', density=True)
plt.title('Distribution of Floating Prediction Errors')
plt.xlabel('Prediction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# PSI Calculation Function
def calculate_psi(train_data, test_data, bins=10):
    # Create bins for both datasets
    train_bins = np.linspace(min(train_data), max(train_data), bins+1)
    test_bins = np.linspace(min(test_data), max(test_data), bins+1)
    
    # Calculate proportions for each bin
    train_hist, _ = np.histogram(train_data, bins=train_bins, density=True)
    test_hist, _ = np.histogram(test_data, bins=test_bins, density=True)
    
    # Avoid division by zero and compute PSI for each bin
    psi = np.sum((test_hist - train_hist) * np.log((test_hist + 1e-10) / (train_hist + 1e-10)))
    return psi

# Call the function to calculate PSI
psi_value = calculate_psi(column_2_values2, column_2_values1, bins=10)

# Output the PSI result
print(f"PSI Value: {psi_value}")

# Interpret the PSI value (commonly accepted thresholds):
# - PSI < 0.1: No drift
# - 0.1 <= PSI < 0.2: Minor drift
# - 0.2 <= PSI < 0.3: Moderate drift
# - PSI >= 0.3: Severe drift
if psi_value < 0.1:
    print("No Drift Detected (PSI < 0.1)")
elif psi_value < 0.2:
    print("Minor Drift Detected (0.1 <= PSI < 0.2)")
elif psi_value < 0.3:
    print("Moderate Drift Detected (0.2 <= PSI < 0.3)")
else:
    print("Severe Drift Detected (PSI >= 0.3)")
```
